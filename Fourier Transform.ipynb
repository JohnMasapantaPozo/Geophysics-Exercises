**Problem 3 - EXERCISE 3**

Make a harmonic signal on the form s(t) = A cos(ωt) = A cos(2πft), where the amplitude is A = 0.5 and the frequency is f=50Hz. Make a trace with 1000 samples and sampling interval 1 ms. Plot the signal.

Fourier transform the signal by FFT and plot the amplitude spectrum. Make a sum of different harmonic signals, and observe the corresponding components in the amplitude spectrum

</p>

```python
#Importing Libraries

import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft, fftfreq, ifft
```

```python
#Setup

n = 1000                          #Number of points
f1 = 50                           #Frequencies
f2 = 80
a1 = 0.5
a2 = 0.7

#Creating individual Signals 
Nt = 1000                         #Number of samples
dt = 0.001                        #Sampling Interval [s]
t = np.linspace(0, (Nt-1)*dt, Nt)
y1 = a1 * np.cos(f1 * 2.0 * np.pi * t)
y2 = a2 * np.cos(f2 * 2.0 * np.pi * t)

#Composed Signal
yf = y1 + y2
```

```python
#Ploting the signal

plt.figure(1)
plt.plot(t*1000, yf)
plt.xlabel('Time [ms]'); plt.ylabel('Amplitude'); plt.title('Composed Signal')
plt.legend(["Sinusoidal Full Signal"], loc = 'upper left')
plt.xlim([0, 150])
plt.grid(True)
```

```python
#Fourier Transform

freqs = fftfreq(n)                       #Creating all the necessary frequencies
mask = freqs > 0                         #Ignoring half of the values since they are complex conjugates of the others

#FFT and power spectra calculations 
fft_vals = fft(yf)
#print(fft_vals)
```

```python
#Ploting the raw values of the FFT

#plt.figure(2)
plt.plot(freqs, fft_vals)
plt.xlabel('Frequency [mHz]'); plt.ylabel('Amplitude'); plt.title('Raw Amplitude Spectrum')
plt.legend(["Raw FFT values"], loc = 'upper left')
plt.xlim([-0.2, 0.2])

plt.grid(True)
```

```python
#True theoretical FFT 

fft_theo = 2.0 * np.abs(fft_vals/n)

#plt.figure(3)
plt.plot(1000*freqs[mask], fft_theo[mask])
plt.xlabel('Frequency [Hz]'); plt.ylabel('Amplitude'); plt.title('True Amplitude Spectrum')
plt.legend(["True FFT values"], loc = 'upper left')
plt.xlim([-150, 150])
plt.grid(True)

print("""Here, thought out the Fourier Transform we can observe 
the amplitudes of the two hamonic waves that make up wave.

Amplitude1 = 0.7 and Amplitude2 = 0.5
Frequency1 = 50 [Hz] and Frequency2 = 80[Hz]
""")

plt.show()
```
